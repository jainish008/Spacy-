```python
import spacy
from spacy import displacy
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
doc = nlp("Facebook is changing its corporate name to Meta. Its flagship social network will keep the Facebook name")
for token in doc:
    print(token.text)
```

```python
displacy.serve(doc, style="dep")
```

```python
displacy.serve(doc, style="ent")
```
